```python
# variables which can't be handled by converting them from string to numerics are called CATEGORICAL VARIABLES

                # types of categorical variables are
     
                #1. NOMINAL - don't have numeric ordering like(male, female)(reg, green, blue)
                #2. ORDINAL - have some sense of order to them like(high, medium, low)(graduate, masters, phd)
```

```python
# for nominals we use - ONE HOT ENDCODING
        # here we CREATE new columns for each categories and ASSIGN A BINARY VALUE(1) WHEN IT'S TRUE
        # these extra variables are called DUMMY VARIABLES
```

```python
import pandas as pd
df = pd.read_csv("D:/coding/Python/newHomeprices.csv")
df
```

```python
# create dummy variables
dummies = pd.get_dummies(df.town)
dummies
```

```python
# concatinate original with dummies(which)& (how)
merged = pd.concat([df,dummies], axis='columns')
merged
```

```python
# now since we have dummny variables, there is no use of texted cols so DROP THEM
# and drop of of dummy varibale bc --> "DUMMY VARIBLE TRAP" <-- which is when we can derive one from rest of variables, then its called multi-colinear & they creates the trap  
```

```python
final = merged.drop(['town','west windsor'], axis = 'columns')
final
```

```python
# now give inputs for training, (X && Y)
x = final.drop('price', axis='columns')
x
```

```python
y = final.price
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x,y) #because x is a DataFrame with feature names, while x.values is only values, without feature names.
```

```python
# for area of 2800 in robinsville
model.predict([[2800,0,1]])
```

```python
# calcuates predicted value for all -- comapres with actual and calculates score
model.score(x,y)
```
